```python
import json
import tweepy

# Load credentials from json file
with open("/home/cpg/.ssh/twitter_credentials.json", "r") as file:  
    creds = json.load(file)

auth = tweepy.OAuthHandler(creds['CKEY'], creds['CSEC'])
auth.set_access_token(creds['ATOK'], creds['ASEC'])
api = tweepy.API(auth)
```

```python
tweets = []
feeds = ['elonmusk', 'realDonaldTrump', 'marwilliamson']
for feed in feeds:
    stuff = api.user_timeline(screen_name = feed, count = 10, include_rts = True) # get last 10 tweets
    for status in stuff:
        link = "https://twitter.com/" + feed + "/status/" + str(status._json["id"])
        tweets.append([feed, status._json["text"], link, status._json["created_at"]])
```

```python
import pandas as pd
tweets = pd.DataFrame(tweets).sort_values(by=[3], ascending=False).reset_index(drop=True)
tweets

#TODO: Convert Timestamps into a Standardized Timezone
```
